```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("C:/Users/RAJKUMAR/Downloads/Python.xlsx")
df.head()
```

```python
# Filter data for years 2017–2025
df = df[(df['year'] >= 2017) & (df['year'] <= 2025)]
```

```python
df_filtered = df[df['CTYNAME'].isin(['China'])]
print(df_filtered['CTYNAME'].unique())
```

```python
import_months = ['IJAN', 'IFEB', 'IMAR', 'IAPR', 'IMAY', 'IJUN',
                 'IJUL', 'IAUG', 'ISEP', 'IOCT', 'INOV', 'IDEC']
export_months = ['EJAN', 'EFEB', 'EMAR', 'EAPR', 'EMAY', 'EJUN',
                 'EJUL', 'EAUG', 'ESEP', 'EOCT', 'ENOV', 'EDEC']
df['Total_Imports'] = df[import_months].sum(axis=1)
df['Total_Exports'] = df[export_months].sum(axis=1)
df['Trade_Balance'] = df['Total_Exports'] - df['Total_Imports']
```

```python
df[import_months + export_months].round(2)
```

```python
print(df[import_months].iloc[0].sum())  # This will sum the first row(imports)
print(df[export_months].iloc[0].sum())  # This will sum the first row(exports)
```

```python
print(df[['Total_Imports', 'Total_Exports', 'Trade_Balance']].head())
```

```python
df['Total_Imports'] = df[import_months].fillna(0).sum(axis=1)# Sum of imports columns, filling NaNs with 0(horizontally)
df['Total_Exports'] = df[export_months].fillna(0).sum(axis=1) # Sum of export columns, filling NaNs with 0
df['Trade_Balance'] = df['Total_Exports'] - df['Total_Imports']# Trade balance is exports minus imports
print(df[['Total_Imports', 'Total_Exports', 'Trade_Balance']].head())# Show first few rows of the result
```

```python
# Group the data by 'year' and 'CTYNAME', then sum the 'Total_Imports' and 'Total_Exports' for each group
yearly = df.groupby(['year', 'CTYNAME'])[['Total_Imports', 'Total_Exports']].sum().reset_index()
print(yearly.head()) 
```

```python
china = yearly[yearly['CTYNAME'] == 'China']

import matplotlib.pyplot as plt

plt.plot(china['year'], china['Total_Imports'], label='Imports from China')
plt.plot(china['year'], china['Total_Exports'], label='Exports to China')
plt.title('U.S. Trade with China (2017–2025)')
plt.xlabel('Year')
plt.ylabel('Trade Value in Dollars')
plt.legend()
plt.grid(True)
plt.show()
```

```python
china = yearly[yearly['CTYNAME'] == 'China'].copy()
china['Trade_Balance'] =  china['Total_Exports'] - china['Total_Imports']
```

```python
plt.bar(china['year'], china['Trade_Balance'], color='skyblue')
plt.axhline(0, color='black', linestyle='--')
plt.title('U.S. Trade Balance with China (2017–2025)')
plt.xlabel('Year')
plt.ylabel('Trade Balance')
plt.show()
```

```python

# 1. Sort and group by country
yearly = yearly.sort_values(by=['CTYNAME', 'year'])

# 2. Calculate YoY changes per country
yearly['Import_YoY_Change'] = yearly.groupby('CTYNAME')['Total_Imports'].pct_change() * 100
yearly['Export_YoY_Change'] = yearly.groupby('CTYNAME')['Total_Exports'].pct_change() * 100

# 3. Filter only China
china = yearly[yearly['CTYNAME'] == 'China']

# 4. Drop the first year (2017) since it has no previous year to compare
china = china[china['year'] > 2017]

# 5. Print clean table
print(china[['year', 'Total_Imports', 'Import_YoY_Change', 'Total_Exports', 'Export_YoY_Change']])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(china['year'], china['Import_YoY_Change'], label='Import % Change', marker='o')
plt.plot(china['year'], china['Export_YoY_Change'], label='Export % Change', marker='o')
plt.axhline(0, color='gray', linestyle='--')
plt.title('Year-over-Year Change in U.S.–China Trade')
plt.xlabel('Year')
plt.ylabel('% Change')
plt.legend()
plt.grid(True)
plt.show()
```
